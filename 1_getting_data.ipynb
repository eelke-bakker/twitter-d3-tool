### Get tweets from users' timelines

```python
address = "data/bot_classification_data.csv"

data = pd.read_csv(address)
data.head()
```

```python
usernames = data["screen_name"].values
```

```python
# loop through usernames, get their tweet data, and save to file

for username in usernames:
    command = "twint -u " + username + " -o tweets.csv --csv --limit 500"
    subprocess.call(command, shell=True)
```
