```python
import pandas as pd
import json
import re
```

### Flatten dictionary columns for current twitter data

```python
address2 = "replies_clean - Copy.csv"

data2 = pd.read_csv(address2)
data2.head()
```

```python
data2.info()
```

```python
df_list = [data2, pd.json_normalize(data2["entities"].apply(eval)), pd.json_normalize(data2["metadata"].apply(eval)), pd.json_normalize(data2["user"].apply(json.loads))]
```

```python
concatenated = pd.concat(df_list, axis=1)
```

```python
concatenated.head()
```

```python
concatenated.info()
```

```python
concatenated["text"] = concatenated["text"].apply(lambda x: re.sub("(@\w+)", "", x))
```

```python
concatenated.drop(columns=['id', 'id_str', 'entities', 'metadata', 'user', 'in_reply_to_status_id_str', 'in_reply_to_user_id_str', 'author', 'id_str.1'], inplace=True)
```
