```python
import pandas as pd
import json
```

### Flatten dictionary columns for current twitter data

```python
address2 = "replies_clean.csv"

data2 = pd.read_csv(address2)
data2.head()
```

```python
data2.info()
```

```python
data2["entities"] = data2["entities"].apply(lambda x: x.replace("'", "\""))
data2["metadata"] = data2["metadata"].apply(lambda x: x.replace("'", "\""))
```

```python
df_list = [data2, pd.json_normalize(data2["entities"].apply(json.loads)), pd.json_normalize(data2["metadata"].apply(json.loads)), pd.json_normalize(data2["user"].apply(json.loads))]
```

```python
concatenated = pd.concat(df_list, axis=1)
```
